```python
df_ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Cleaned Data', 'orders_products_merged.pkl'))
```

```python
# Creating subset

df = df_ords_prods_merged[:1000000]
```

```python
# Creating price categories

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] >= 5) and (row['prices'] <=15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data' 
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df.value_counts()
```

```python
# checking highest priced product in subset
df['prices'].max()
```

    # Creating price categories using .loc function


```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <=5, 'price_range_loc'] = 'Low_range_product'
```

```python
# finding the busiest day
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Finding busiest day by loop

result = []
for value in df_ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

    # Task 4 - Make a new column that categorises the busiest time of day

```python
# Checking busiest/slowest time day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

    # 
At this point is moved these numbers to an Excel sheet in order to run some calulations. 
I wanted to find the average and standard deviation of these totals so that I could categorise the 'busiest hours/'slowest hours' by what is one standard deviation above and below our average. I took this approach because I wanted the time brackets to be relative to the average workload. 

```python
# Loop to create and populate 'workload' column

workload = []

for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10,11,13,14,15]:
        workload.append('Most orders')
    elif value in [0,6,22,23]:
        workload.append('Fewest orders')
    else:
        workload.append('Average orders')
```

```python
# Checking loop success
workload
```

```python
df_ords_prods_merged['busiest_period_of_day'] = workload
```

```python
# Checking column creation
df_ords_prods_merged.head()
```
