#  1. Key Questions 4 - Creating a pie chart to show the distribution of price range products

# 2. Key Question 5a - What is the distribution among users in regards to their loyalty?

# 3. Key Questions 5b - What are the differences in ordering habits based on customer's loyalty status 

```python
# Cross referencing loyalty flag and spending flag columns
```

```python
loyalty_avg_price = pd.crosstab(df['loyalty_flag'], df['spending_flag'], dropna = False)
```
